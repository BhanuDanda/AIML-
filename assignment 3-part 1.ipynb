```python
print(1.4)
from abc import ABC, abstractmethod
import copy

# Define a generic Game class (Superclass)
class Game(ABC):
    @abstractmethod
    def actions(self, state):
        pass

    @abstractmethod
    def result(self, state, action):
        pass

    @abstractmethod
    def is_terminal(self, state):
        pass

    @abstractmethod
    def utility(self, state, player):
        pass


# Subclass for TicTacToe game
class TicTacToe(Game):
    def __init__(self):
        # Initial state is a 3x3 board with empty cells
        self.initial_state = [['' for _ in range(3)] for _ in range(3)]
        self.player_x = 'X'
        self.player_o = 'O'

    # Return the available actions in the current state (i.e., empty cells)
    def actions(self, state):
        return [(i, j) for i in range(3) for j in range(3) if state[i][j] == '']

    # Apply the action and return the new state
    def result(self, state, action):
        new_state = copy.deepcopy(state)
        i, j = action
        current_player = self.current_player(state)
        new_state[i][j] = current_player
        return new_state

    # Check if the game is over
    def is_terminal(self, state):
        # Check if there's a winner
        if self.check_winner(state, self.player_x) or self.check_winner(state, self.player_o):
            return True

        # Check if the board is full (no empty spaces)
        return all(state[i][j] != '' for i in range(3) for j in range(3))

    # Return the utility of the state for a player
    def utility(self, state, player):
        if self.check_winner(state, player):
            return 1  # Player wins
        elif self.check_winner(state, self.player_o if player == self.player_x else self.player_x):
            return -1  # Opponent wins
        else:
            return 0  # Draw or non-terminal state

    # Helper method to check current player
    def current_player(self, state):
        x_count = sum(row.count(self.player_x) for row in state)
        o_count = sum(row.count(self.player_o) for row in state)
        return self.player_x if x_count == o_count else self.player_o

    # Helper method to check if a player has won
    def check_winner(self, state, player):
        # Check rows, columns, and diagonals
        return (
            any(all(state[i][j] == player for j in range(3)) for i in range(3)) or  # Rows
            any(all(state[i][j] == player for i in range(3)) for j in range(3)) or  # Columns
            all(state[i][i] == player for i in range(3)) or  # Main diagonal
            all(state[i][2 - i] == player for i in range(3))  # Anti diagonal
        )

# Example of how to use the TicTacToe class
if __name__ == "__main__":
    game = TicTacToe()
    state = game.initial_state
    print("Initial State:")
    for row in state:
        print(row)

    # Example of actions
    print("\nAvailable Actions:")
    print(game.actions(state))

    # Example of result (X makes a move at (0, 0))
    new_state = game.result(state, (0, 0))
    print("\nState After X moves to (0, 0):")
    for row in new_state:
        print(row)

    # Example of is_terminal and utility
    print("\nIs Terminal:", game.is_terminal(new_state))
    print("Utility for X:", game.utility(new_state, 'X'))
```

```python
print(1.6)
import functools
import copy

# Define a cache for computationally expensive functions
cache = functools.lru_cache(maxsize=10**6)

class Game:
    def __init__(self, initial_state):
        """
        Initialize the game with the initial state.
        """
        self.initial_state = initial_state

    def actions(self, state):
        """
        Return a collection of the allowable moves from this state.
        This method should be implemented in a subclass.
        """
        raise NotImplementedError("Subclasses must implement this method.")

    def result(self, state, move):
        """
        Return the state that results from making a move from a state.
        This method should be implemented in a subclass.
        """
        raise NotImplementedError("Subclasses must implement this method.")

    def is_terminal(self, state):
        """
        Return True if this is a final state for the game.
        """
        return not self.actions(state)

    def utility(self, state, player):
        """
        Return the value of this final state to the player.
        This method should be implemented in a subclass.
        """
        raise NotImplementedError("Subclasses must implement this method.")

    def reset(self):
        """
        Reset the game to the initial state.
        """
        return self.initial_state

    def current_player(self, state):
        """
        Determine the current player based on the state.
        """
        raise NotImplementedError("Subclasses must implement this method.")

class TicTacToe(Game):
    def __init__(self):
        """
        Initialize the Tic-Tac-Toe game with the default initial state.
        """
        initial_state = tuple(' ' for _ in range(9))  # 3x3 board as a tuple
        super().__init__(initial_state)

    def actions(self, state):
        """
        Return a list of valid moves (empty positions) for the current player.
        """
        return [i for i, cell in enumerate(state) if cell == ' ']

    def result(self, state, move):
        """
        Return the new state after applying the move.
        """
        new_state = list(state)
        current_player = self.current_player(state)
        new_state[move] = current_player
        return tuple(new_state)

    def is_terminal(self, state):
        """
        Check if the game has ended.
        """
        return self.check_winner(state) is not None or ' ' not in state

    def utility(self, state, player):
        """
        Return the utility of the final state for the player.
        """
        winner = self.check_winner(state)
        if winner == player:
            return 1  # Win
        elif winner is None:
            return 0  # Draw
        else:
            return -1  # Loss

    def current_player(self, state):
        """
        Determine the current player based on the state.
        """
        x_count = state.count('X')
        o_count = state.count('O')
        return 'X' if x_count == o_count else 'O'

    def check_winner(self, state):
        """
        Determine if there is a winner in the game.
        """
        winning_combinations = [
            (0, 1, 2), (3, 4, 5), (6, 7, 8),  # Rows
            (0, 3, 6), (1, 4, 7), (2, 5, 8),  # Columns
            (0, 4, 8), (2, 4, 6)               # Diagonals
        ]
        for a, b, c in winning_combinations:
            if state[a] == state[b] == state[c] != ' ':
                return state[a]
        return None

def play_game(game, strategies, verbose=False):
    """
    Play a turn-taking game. 'strategies' is a {player name: function} dict,
    where function(state, game) is used to get the player's move.
    """
    state = game.reset()
    while not game.is_terminal(state):
        player = game.current_player(state)
        move = strategies[player](state, game)
        state = game.result(state, move)

        if verbose:
            print(f'Player {player} move: {move}')
            print(state)

    if verbose:
        print("Game Over")

    return state

# Example strategies
def random_strategy(state, game):
    import random
    return random.choice(game.actions(state))

def main():
    game = TicTacToe()

    strategies = {
        'X': random_strategy,
        'O': random_strategy
    }

    final_state = play_game(game, strategies, verbose=True)
    print("Final State:", final_state)

if __name__ == "__main__":
    main()

```
